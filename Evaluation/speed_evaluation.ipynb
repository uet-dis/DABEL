```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the bar chart in milliseconds
scenarios = ['TGAN', 'FDM', 'CFM']
xgb = [516.484260559082, 50.0340461730957, 52.86812782287598]  # converted from seconds to ms
et = [129.4846534729004, 48.14314842224121, 36.16690635681152]   # converted from seconds to ms
rf = [20.90167999267578, 34.35063362121582, 54.3670654296875]  # converted from seconds to ms
seq_voting = [671.9682693481445, 98.81591796875, 123.3980655670166]  # in milliseconds
par_voting = [589.3659591674805, 76.8594741821289, 81.79378509521484]  # in milliseconds

# Bar width
bar_width = 0.12
# Spacing between bars within the same scenario
bar_spacing = 0.02
# Reduced spacing between different scenarios
scenario_spacing = 0.4

# Calculate the number of bars per scenario
num_bars_per_scenario = 5

# Positions of the bars on the x-axis with space between bars
r1 = np.arange(len(scenarios)) * (bar_width * num_bars_per_scenario + scenario_spacing)
r2 = [r + bar_width + bar_spacing for r in r1]
r3 = [r + (bar_width + bar_spacing) * 2 for r in r1]
r4 = [r + (bar_width + bar_spacing) * 3 for r in r1]
r5 = [r + (bar_width + bar_spacing) * 4 for r in r1]

# Create the plot
plt.figure(figsize=(10, 8))  # Adjust the figure size as needed
bars1 = plt.bar(r1, xgb, color='orange', width=bar_width, edgecolor='grey', label='XGB')
bars2 = plt.bar(r2, et, color='blue', width=bar_width, edgecolor='grey', label='ET')
bars3 = plt.bar(r3, rf, color='purple', width=bar_width, edgecolor='grey', label='RF')
bars4 = plt.bar(r4, seq_voting, color='red', width=bar_width, edgecolor='grey', label='SeqEns')
bars5 = plt.bar(r5, par_voting, color='green', width=bar_width, edgecolor='grey', label='ParaEns')

# Adding the labels
plt.xlabel('Scenarios S3', fontweight='bold')
plt.ylabel('Time (ms)', fontweight='bold')

# Calculate xticks positions dynamically based on the number of bars per scenario
xticks_positions = [r + (bar_width * (num_bars_per_scenario // 2) + bar_spacing * (num_bars_per_scenario // 2)) for r in r1]
plt.xticks(xticks_positions, scenarios)

plt.title('Time Evaluation of DAELID-based Intrusion Detection (ms)')
plt.legend()

# Adding the values on top of each bar with inclined style
def add_value_labels(bars):
    for bar in bars:
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2, yval + 10, '{:.2f}'.format(yval), ha='center', va='bottom', rotation=45)

add_value_labels(bars1)
add_value_labels(bars2)
add_value_labels(bars3)
add_value_labels(bars4)
add_value_labels(bars5)

# Extend the length of the y-axis
plt.ylim(0, 800)  # Adjust the upper limit as needed

# Show the plot
plt.tight_layout()
plt.show()
```
